```python
import joblib
import pandas as pd

model = joblib.load("models/trained_model.pkl")
with open("models/base_year.txt", "r") as f:
    base_year = int(f.read())

# Manually test predictions
for year in [2025, 2026, 2027, 2028, 2029]:
    year_since_start = year - base_year
    test = {
        "year_since_start": year_since_start,
        "area_ha": 10,
        "crop_Bambara groundnut": 1,  # assuming that's how it's encoded
    }

    # Add missing features with 0
    for feat in model.feature_names_in_:
        if feat not in test:
            test[feat] = 0

    test_df = pd.DataFrame([test])
    pred = model.predict(test_df)[0]
    print(f"Year {year} â†’ {round(pred, 2)} tons/ha")
```
